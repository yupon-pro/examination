```python

def combine_lines(lines, initial_lines):
  combined_lines = []
  
  target_lines = [*initial_lines, *lines]
  num_lines = len(target_lines)

  for i in range(num_lines):
    for j in range(i+1, num_lines):
      former_line = target_lines[i]
      latter_line = target_lines[j]

      former_start = former_line[0]
      former_end = former_line[-2]
      latter_start = latter_line[0]
      latter_end = latter_line[-2]

      if former_start == former_end or latter_start == latter_end:
        continue

      
      if former_end == latter_start:

        if former_start == latter_end:
          former_rest_stations = set(former_line[1:-2])
          latter_rest_stations = set(latter_line[1:-2])
        else:
          former_rest_stations = set(former_line[:-2])
          latter_rest_stations = set(latter_line[1:-1])

        intersection_flag = len(former_rest_stations & latter_rest_stations) >= 1
        duplication_flag = former_end in latter_rest_stations or latter_start in former_rest_stations

        if intersection_flag or duplication_flag:
          continue

        distance = former_line[-1] + latter_line[-1]
        combined_line = [*former_line[:-2], *latter_line[:-1], distance]
        combined_lines.append(combined_line)

      if former_start == latter_end:
        if former_end == latter_start:
          former_rest_stations = set(former_line[1:-2])
          latter_rest_stations = set(latter_line[1:-2])
        else:
          former_rest_stations = set(former_line[1:-1])
          latter_rest_stations = set(latter_line[:-2])

        intersection_flag = len(former_rest_stations & latter_rest_stations) >= 1
        duplication_flag = former_start in latter_rest_stations or latter_end in former_rest_stations

        if intersection_flag or duplication_flag:
          continue

        distance = former_line[-1] + latter_line[-1]
        combined_line = [*latter_line[:-2], *former_line[:-1],  distance]
        combined_lines.append(combined_line)

  return combined_lines


def repeat_combination(lines, initial_lines):
  prev_length = len(lines)
  integrated_lines = combine_lines(lines, initial_lines)
  if prev_length == len(integrated_lines):
    return integrated_lines
  else:
    return repeat_combination(integrated_lines, initial_lines)
  
```
